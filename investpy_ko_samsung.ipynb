```python
from investpy.stocks import get_stock_historical_data as gshd
from datetime import datetime, timedelta
import pandas as pd
import os
```

```python
samsung = gshd('005930', 'south korea', '01/01/2018', '16/04/2020')
samsung
```

```python
articles = pd.read_csv('.\\articles_ko_samsung.csv', encoding='utf-8', sep=',')
articles
```

```python
articles_dates = list(pd.to_datetime(articles.date))
kospi_dates = list(samsung.index)
```

```python
data = [[], []]
columns = ['date', 'updown']
for date in articles_dates:
    if date in kospi_dates:
        idx = kospi_dates.index(date)
        if idx != len(kospi_dates) - 1:
            diff = samsung.iloc[idx, 3] - samsung.iloc[idx, 0]
            if diff >= 0:
                data[1].append(1)
            else:
                data[1].append(0)
            data[0].append(date)
        else:
            data[1].append(1)
            data[0].append(date)
    else:
        for compare in kospi_dates:
            if abs(compare - date) <= timedelta(days=1):
                idx = kospi_dates.index(compare)
                break
        if idx != 0:
            diff = samsung.iloc[idx, 3] - samsung.iloc[idx - 1, 0]
            if diff >= 0:
                data[1].append(1)
            else:
                data[1].append(0)
            data[0].append(date)
        else:
            data[1].append(1)
            data[0].append(date)
```

```python
new_df = pd.DataFrame(data).T
new_df.columns = columns
```

```python
new_df.set_index('date', inplace=True)
new_df.to_csv('.\\label_ko_samsung.csv', encoding='utf-8', sep=',')
```
