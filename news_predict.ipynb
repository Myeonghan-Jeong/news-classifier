```python
from tensorflow.keras.models import load_model
```

```python
ko_model = load_model('.\\best_model_ko.h5')
```

```python
from konlpy.tag import Okt
```

```python
tokenizer = Okt()
```

```python
import re
```

```python
words = {}
title = '[충격][속보] 삼성전자를 향한 세계 유수의 회사들의 구애!'
title = re.sub(r'[!"#$%&()*+.,-/:;=?@[\]^_`{|}~\'0-9a-zA-Z·…]', '', title)
title = re.sub(r'[ ]+', '', title)
tokens = tokenizer.morphs(title)
for token in tokens:
    words[token] = 1 if token not in words else words[token] + 1
```

```python
word_to_idx = {}
for key, val in sorted(words.items(), key=lambda x: -x[1]):
    word_to_idx[key] = len(word_to_idx) + 1
    if len(word_to_idx) >= 5000:
        break
```

```python
X = []
title = '[충격][속보] 삼성전자를 향한 세계 유수의 회사들의 구애!'
title = re.sub(r'[!"#$%&()*+.,-/:;=?@[\]^_`{|}~\'0-9a-zA-Z·…]', '', title)
title = re.sub(r'[ ]+', '', title)
tokens = tokenizer.morphs(title)
raw = []
for token in tokens:
    num = word_to_idx[token] if token in word_to_idx else 0
    raw.append(num)
if len(raw) < 30:
    raw += [0] * (30 - len(raw))
X.append(raw)
```

```python
import numpy as np
```

```python
X = np.array(X)
```

```python
ko_model.predict(X)
```
